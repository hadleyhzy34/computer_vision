### convolution process

```python
import tensorflow as tf
import numpy as np
```

## input source

```python
x = np.array([[1,2,3,7],[6,3,4,2],[5,0,5,0],[4,1,8,4]])
```

## kernel

```python
kernel_filter = np.array([[1,-1],[1,-1]])
```

## convolution process

```python
tensor_x = tf.constant(x, tf.float32)
tensor_k = tf.constant(kernel_filter, tf.float32)

# be careful that convolution input must be both 4 dimensional
# definition of stride and padding:https://towardsdatascience.com/covolutional-neural-network-cb0883dd6529
# padding difference: 'SAME' or 'VALID', https://stackoverflow.com/questions/37674306/what-is-the-difference-between-same-and-valid-padding-in-tf-nn-max-pool-of-t
tensor_res = tf.nn.convolution(tf.reshape(tensor_x, [1, 4, 4, 1]), tf.reshape(tensor_k, [2, 2, 1, 1]), strides= 2, padding='VALID')

# sess = tf.Session()
# print(sess.run(tensor_res))
print(tensor_res)
```

```python
tensor_res+1
```

# another example

```python
x = np.array([[1,2,3,7],[6,3,4,2]])
k = np.array([[-0.0308,-0.4722],[-0.3021,-0.0464]])
tensor_x = tf.constant(x,tf.float32)
tensor_k = tf.constant(k,tf.float32)
tensor_res = tf.nn.convolution(tf.reshape(tensor_x,[1,2,4,1]),
tf.reshape(tensor_k,[2,2,1,1]), strides=2, padding='VALID')
print(tensor_res)
```
